```python
#Q2
def factorial(n):
  if(n==0 or n==1) :
    return 1
  return n*factorial(n-1)

print(factorial(5))
```

```python
#Q5
# 1. Input Student Details
student_name = input("Enter Student Name: ")
roll_number = input("Enter Roll Number: ")

# 2. Input Marks for 3 Subjects (Assuming max marks per subject is 100)
print("\nEnter marks obtained (out of 100):")
subject1 = float(input("Enter marks for Math: "))
subject2 = float(input("Enter marks for Science: "))
subject3 = float(input("Enter marks for English: "))

# 3. Calculate Total and Percentage
total_marks = subject1 + subject2 + subject3
max_marks = 300
percentage = (total_marks / max_marks) * 100

# 4. Display Grade using if-elif-else structures
if percentage >= 90 and percentage <= 100:
    grade = "A+"
elif percentage >= 80:
    grade = "A"
elif percentage >= 70:
    grade = "B"
elif percentage >= 60:
    grade = "C"
elif percentage >= 50:
    grade = "D"
elif percentage >= 0 and percentage < 50:
    grade = "Fail"
else:
    grade = "Invalid Marks Calculated"

# 5. Display the Report Card
print("\n" + "="*30)
print("         STUDENT REPORT CARD      ")
print("="*30)
print(f"Name        : {student_name}")
print(f"Roll Number : {roll_number}")
print(f"Total Marks : {total_marks} / {max_marks}")
print(f"Percentage  : {percentage:.2f}%")
print(f"Grade       : {grade}")
print("="*30)
```
