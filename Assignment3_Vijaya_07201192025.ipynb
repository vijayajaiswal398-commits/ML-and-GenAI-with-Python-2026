```python
#Q3
def reverse(n):
  rev=0
  while(n!=0):
    rem=n%10
    rev=10*rev+rem
    n//=10
  return rev

reverse(123)
```

```python
#Q4
def count(n):
  ans=0
  while(n!=0):
    n//=10
    ans+=1
  return ans

count(67)
```

```python
#Q5
def palindrome(n):
  rev=0
  num=n
  while(n!=0):
    rem=n%10
    rev=10*rev+rem
    n//=10
  return rev==num

palindrome(1221)
```

```python
#Q6
def fib(n):
  for i in range(n):
    if(i==0 or i==1):
      print(i)
    else:
      print(i-1+i-2)

fib(10)
```

```python
#Q7
# 1. Define the functions
def add(x, y):
    return x + y

def subtract(x, y):
    return x - y

def multiply(x, y):
    return x * y

def divide(x, y):
    return x / y

# 2. User selects operation
print("1: Add | 2: Subtract | 3: Multiply | 4: Divide")
choice = input("Select an operation (1/2/3/4): ")

# 3. Get the numbers
num1 = float(input("Enter the first number: "))
num2 = float(input("Enter the second number: "))

# 4. Program performs calculation and displays result
if choice == '1':
    print("Result:", add(num1, num2))
elif choice == '2':
    print("Result:", subtract(num1, num2))
elif choice == '3':
    print("Result:", multiply(num1, num2))
elif choice == '4':
    print("Result:", divide(num1, num2))
else:
    print("Invalid choice.")

```

```python
#Q10
# Ask the user for two numbers
num1 = float(input("Enter the numerator: "))
num2 = float(input("Enter the denominator: "))

# The 'try' block contains the code that might cause an error
try:
    result = num1 / num2
    print("The result is:", result)

# The 'except' block catches the specific division by zero error
except ZeroDivisionError:
    print("Error: You cannot divide a number by zero!")
```

```python
#Q11
# Define the class blueprint
class Student:
    # The __init__ method runs automatically when a new student is created
    def __init__(self, student_name, student_marks):
        self.name = student_name
        self.marks = student_marks

    # A simple function to display the student's info
    def display_info(self):
        print("Student Name:", self.name)
        print("Marks:", self.marks)

# --- Testing the class ---

# Create a new student object using the blueprint
student1 = Student("Vijaya", 92)
student2 = Student("Vanshika", 85)

# Display their details
student1.display_info()
print("---")
student2.display_info()
```
